```python
df=pd.DataFrame(datasets.data,columns=datasets.feature_names)
df.head()
```

```python
df.describe()
```

```python
X=df
y=datasets.target
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)

X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)  
#  retain 95 percent of use full information  
#  and then creat new dimension
#  and whatever features it come of with is happier

X_pca=pca.fit_transform(X)
X_pca.shape
```

```python
#  here the resolution is reduced to 29  
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
train_test_split(X_pca,y,test_size=0.2,random_state=10)
```

```python
X_train_pca,X_test_pca,y_train,y_test=train_test_split(X_pca,y,test_size=0.2,random_state=10)
```

```python
modelpca=LogisticRegression(max_iter=1000)
modelpca.fit(X_train_pca,y_train)
modelpca.score(X_test_pca,y_test)
```

```python
# if we take only 2 componenets useful components
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X)
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca,X_test_pca,y_train,y_test=train_test_split(X_pca,y,test_size=0.2,random_state=10)
modelpca1=LogisticRegression(max_iter=1000)
modelpca1.fit(X_train_pca,y_train)
modelpca1.score(X_test_pca,y_test)
```
